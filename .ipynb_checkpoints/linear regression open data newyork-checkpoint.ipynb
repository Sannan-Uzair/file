# Model Building 

```python
df.drop(['name', 'id', 'host_name','last_review'], axis=1, inplace=True)
df['reviews_per_month'] = df['reviews_per_month'].replace(np.nan, 0)
```

```python
df.head()
```

```python
# Label Encoding

le = LabelEncoder()
le.fit(df['neighbourhood_group'])
df['neighbourhood_group'] = le.transform(df['neighbourhood_group'])

le.fit(df['neighbourhood'])
df['neighbourhood'] = le.transform(df['neighbourhood'])

le.fit(df['room_type'])
df['room_type'] = le.transform(df['room_type'])

df.sort_values(by='price', ascending = True, inplace = True)
```

```python
df.head()
```

```python
min_threshold1, max_threshold1 = df.price.quantile([0.05, 0.95])
min_threshold1, max_threshold1

df = df[(df.price < max_threshold1) & (df.price > min_threshold1)]
df.shape
```

```python
X = df.drop('price',axis=1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 101)
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
pred = model1.predict(X_test)
print('Mean squared error: ',np.sqrt(mean_squared_error(y_test, pred)))
print('R2 score: ', r2_score(y_test, pred)*100)
print('Mean absolute erro: ', mean_absolute_error(y_test, pred))
```

```python
res = pd.DataFrame({'Actual Value': np.array(y_test).flatten(),
                   'Predicted Values': pred.flatten()})
res.head()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(pred, y_test)
plt.xlabel('Predictions')
plt.ylabel('Actual')
plt.title('Linear Regression Prediction')
plt.grid(False)
plt.show();
```

```python
gboost = GradientBoostingRegressor()
gboost.fit(X_train, y_train)
gboost.score(X_test, y_test)
```

```python
pred = gboost.predict(X_test)
print('Mean squared error: ',np.sqrt(mean_squared_error(y_test, pred)))
print('R2 score: ', r2_score(y_test, pred)*100)
print('Mean absolute erro: ', mean_absolute_error(y_test, pred))
```
